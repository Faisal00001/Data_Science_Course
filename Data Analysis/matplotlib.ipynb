```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
```

```python
y = [1,4,16,25,59]
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic line plot')
plt.show()
```

```python
# create a customized line plot
x = [1,2,3,4,5]
y = [1,4,16,25,59]
plt.plot(x,y,color='red',linestyle='--')
```

```python
x = [1,2,3,4,5]
y = [1,4,16,25,59]
plt.plot(x,y,color='red',linestyle='-.')
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 16, 25, 59]

plt.plot(x, y, color='red', linestyle='-.', marker='o', linewidth=3, markersize=9)
plt.grid(True)
plt.show()
```
